```python
# 不要顯示瀏覽器
option = webdriver.ChromeOptions()
option.add_argument("headless")

#定義抓取google校正地址的資料 流程
def get_addr(addr):
    addr=addr.replace(" ","")
    addr=addr.replace("　","")
    try:
        if(len(addr)>=8):
            driver = webdriver.Chrome('D:/Users/RT009223/Desktop/chromedriver.exe',chrome_options=option)
            time.sleep(3)

            driver.get('https://www.google.com/search?q='+addr)
            time.sleep(2)
            soup = BeautifulSoup(driver.page_source, 'lxml')
            time.sleep(2)
            try:
                result=soup.find_all('div',class_="desktop-title-content")[0].text[3:]
                print(addr,'第一次成功！')
            except:
                try:
                    result=soup.find_all('a',class_="gL9Hy")[0].text
                    print(addr,'第二次成功！')
                except:
                    result=addr
                    print(addr,'失敗！')
            finally:
                try:
                    driver.close()
                except:
                    pass
            return result
        else:
            print(addr,'地址長度不夠！')
            return ""
    except:
        print('執行錯誤！')
        return '0'
```

```python
# 抓去目前更新的最新位置
def start_index(df_input):
    try:
        return max(df_input[df_input.Done ==1].index)+1
    except:
        return 0
```

```python
# 從哪裡開始更新
idx=start_index(df_input)
idx
```

```python
# 需要更新的總筆數
df_input.shape[0]
```

```python
idx=start_index(df_input)
ans="0"
n_finish=0
total_wrong=0
f=0
for i in range(idx,df_input.shape[0]):
    n=0
    while ans == "0":
        ans=get_addr(addr=df_input.Address[i])
        n+=1
        if n>10:
            total_wrong+=1
            continue
    if df_input.new_addr[i] != ans:
        df_input.Done[i]=1
        f=0
    else:
        f+=1
    if f>20:
        break
    df_input.new_addr[i]=ans
    ans="0"
    n_finish+=1
    if n_finish%50==0:
        print('完成數量 : ',n_finish)
        df_input.to_excel('sample01.xlsx',index=0,encoding='utf-8-sig')
    if total_wrong >50:
        break
```
